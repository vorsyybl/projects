```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
raw = pd.read_csv('mw_90.csv')
```

```python
raw.drop(['Unnamed: 0', 'Unnamed: 15', 'Unnamed: 1'], axis=1, inplace=True)
raw.rename(columns={'Game': 'title'}, inplace=True)
raw.drop([0, 1, 5, 6, 8, 9, 11, 12, 13], axis=0, inplace=True)
```

```python
top10 = raw.loc[:, ['title', 'Watch time', 'Average viewers', 'Followers gained']].head(10)
```

```python
top10.plot.bar(x='title', y='Followers gained')
```

```python
top10.plot.pie(y='Followers gained')
```
